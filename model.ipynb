```python
from skimage.color import rgb2lab, lab2rgb, rgb2gray
from skimage.io import imsave
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import numpy as np
import cv2
import os, random
import tensorflow as tf
```

```python
# Loading the images from data folder
# dataset: http://www.ifp.illinois.edu/~vuongle2/helen
X = []
for imagename in os.listdir('data/train/'):
    X.append(img_to_array(load_img('data/train/'+imagename)))
X = np.array(X, dtype=float)

# Split into training and testing dataset
split = int(0.95*len(X))
Xtrain = X[:split]
Xtrain = 1.0/255*Xtrain
```

```python
from keras.callbacks import TensorBoard, ModelCheckpoint
from keras.models import Sequential
from keras.layers import Conv2D, Conv2DTranspose, UpSampling2D, Activation, Dense, Dropout, Flatten, InputLayer
from keras.layers.normalization import BatchNormalization

# Construct convolutional layers
def addConvLayer(model, filter_size, strides=0):
    if strides!=0:
        model.add(Conv2D(filter_size, (3, 3), strides=2, padding='same', activation='relu'))
    else:
        model.add(Conv2D(filter_size, (3, 3), padding='same', activation='relu'))

model = Sequential()

# Input Layer
model.add(Conv2D(64, (3, 3), input_shape=(256, 256, 2), activation='relu', padding='same'))

# Convolutional Layers
addConvLayer(model, 64, strides=2)
addConvLayer(model, 128)
addConvLayer(model, 128, strides=2)
addConvLayer(model, 256)
addConvLayer(model, 256, strides=2)
addConvLayer(model, 512)
addConvLayer(model, 256)
addConvLayer(model, 128)

# Deconvolutional Layers
model.add(UpSampling2D((2, 2)))
addConvLayer(model, 64)
model.add(UpSampling2D((2, 2)))
addConvLayer(model, 32)

# Output Layer
model.add(Conv2D(2, (3, 3), activation='tanh', padding='same'))
model.add(UpSampling2D((2, 2)))

# Compiling the CNN
model.compile(optimizer='rmsprop', loss='mse')
```

```python
## ADD CONTROUR INFORMATION
# reference: https://docs.opencv.org/3.4/d3/dc0/group__imgproc__shape.html#ga17ed9f5d79ae97bd4c7cf18403e1689a
def img_to_cont(batch):
    
    lab_batch = rgb2lab(batch)
    
    res = []
    
    for rimg in lab_batch:

        #== Parameters =======================================================================
        BLUR = 21
        CANNY_THRESH_1 = 10
        CANNY_THRESH_2 = 200
        MASK_DILATE_ITER = 10
        MASK_ERODE_ITER = 10
        MASK_COLOR = 1.0


        #== Processing =======================================================================

        #-- Read image -----------------------------------------------------------------------
        #img = cv2.imread('C:/Users/heyuc/Desktop/New folder/person.png')
        cv2.imwrite('color_img.jpg', rimg[:,:,0])
        cv2.waitKey()
        img = cv2.imread('color_img.jpg')
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
        #print(gray)

        #-- Edge detection -------------------------------------------------------------------
        edges = cv2.Canny(gray, CANNY_THRESH_1, CANNY_THRESH_2)
        edges = cv2.dilate(edges, None)
        edges = cv2.erode(edges, None)

        #-- Find contours in edges, sort by area ---------------------------------------------
        contour_info = []
        contours, _ = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

        # # Previously, for a previous version of cv2, this line was: 
        # #  contours, _ = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        # # Thanks to notes from commenters, I've updated the code but left this note
        for c in contours:
            contour_info.append((
                c,
                cv2.isContourConvex(c),
                cv2.contourArea(c),
            ))
        contour_info = sorted(contour_info, key=lambda c: c[2], reverse=True)
        if contour_info:
            max_contour = contour_info[0]

            #print(max_contour)


            # #-- Create empty mask, draw filled polygon on it corresponding to largest contour ----
            # # Mask is black, polygon is white
            mask = np.zeros(edges.shape)
            cv2.fillConvexPoly(mask, max_contour[0], (255))

            #-- Smooth mask, then blur it --------------------------------------------------------
            mask = cv2.dilate(mask, None, iterations=MASK_DILATE_ITER)
            mask = cv2.erode(mask, None, iterations=MASK_ERODE_ITER)
            mask = cv2.GaussianBlur(mask, (BLUR, BLUR), 0)
            
            #cv2.imshow('mask', mask)
            

            res.append(mask)
            
            
        else:
            # If no contour is detected, we just return blank
            res.append(rimg[:,:,0])
    
    #print(np.stack((res, lab_batch[:,:,:,0]),axis=-1).shape)
    return np.stack((lab_batch[:,:,:,0], res),axis=-1)
```

```python
# Transform image to input data with contour
import matplotlib.pyplot as plt

datagen = ImageDataGenerator(
        shear_range=0.2,
        zoom_range=0.2,
        rotation_range=20,
        horizontal_flip=True)

# Generate training data
batch_size = 10
def image_a_b_gen(batch_size):
    for batch in datagen.flow(Xtrain, batch_size=batch_size):
        lab_batch = rgb2lab(batch)
        X_batch = img_to_cont(batch)
        Y_batch = lab_batch[:,:,:,1:] / 128
        yield (X_batch, Y_batch)

```

```python
# Train the model      
tensorboard = TensorBoard(log_dir=".\log")
trainedmodel = model.fit_generator(image_a_b_gen(batch_size), callbacks=[tensorboard],epochs=500, steps_per_epoch=40)
```
